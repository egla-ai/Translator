```python
# Prepare the data for training
train_texts = train_df["Dholuo"].astype(str).tolist()  # Convert to strings explicitly
train_labels = train_df["Kiswahili"].astype(str).tolist() # Convert to strings explicitly

# Tokenize and prepare model input tensors
train_encodings = tokenizer(train_texts, return_tensors="pt", padding=True, truncation=True)
train_labels_encodings = tokenizer(train_labels, return_tensors="pt", padding=True, truncation=True)
# Reduce Dataset Size for Faster Experimentation (Optional)
train_df = train_df.sample(frac=0.1, random_state=42)  # Use 10% of data for testing

# Batch Size Adjustment
batch_size = 8  # or 16, 32, etc. - Experiment to find the optimal value

# Utilize GPU if Available
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device) # Move the model to the device

# Training loop (simplified example)
for epoch in range(3):  # Increase epochs for better results (but longer training)
    model.train()
    for i in range(0, len(train_encodings["input_ids"]), batch_size):
        input_ids = train_encodings["input_ids"][i:i+batch_size].to(device) # Move data to device
        attention_mask = train_encodings["attention_mask"][i:i+batch_size].to(device) # Move data to device
        labels = train_labels_encodings["input_ids"][i:i+batch_size].to(device) # Move data to device

        # Forward pass (include attention_mask)
        outputs = model(input_ids=input_ids, attention_mask=attention_mask, labels=labels)
        loss = outputs.loss

        # Backward pass and optimization step
        loss.backward()
        optimizer.step()
        optimizer.zero_grad()

    print(f"Epoch {epoch+1}, Loss: {loss.item()}")

print("Training complete.")
```
